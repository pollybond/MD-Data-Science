```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

Визуализация распределений и рассчет мат. ожидания, дисперсии, ассиметрии, эксцесс моментов. Квантили уровня 0,05 и 0,95; 2,5%-ную точку  для найденной случайной величины. Сравнительная диаграмма найденного распределения и распределения по нормальному закону.

1. Анализ данных о весе новорожденных (1960-1967 года, Сан-Франциско, источник: https://www.kaggle.com/datasets/debjeetdas/babies-birth-weight?resource=download)

Чтение данных и визуализация распеределения

```python
file_path = "C:/Users/bonda/OneDrive/Документы/Магистратура МАИ+СБЕР/Теор. вер. и мат. статистика/ЛР1/babies.csv"
data = pd.read_csv(file_path)

# Извлечение столбца с весом новорожденных
weights = data["weight"].dropna()  # Убираем пропуски

# Построение гистограммы
plt.hist(weights, bins=20, density=True, alpha=0.6, color="g", label="Вес новорожденных")

# Оценка параметров нормального распределения
mu_1, std_1 = weights.mean(), weights.std()

# Создание кривой нормального распределения
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_1, std_1)

# Наложение кривой нормального распределения
plt.plot(x, p, "k", linewidth=2, label="Нормальное распределение")

# Оформление графика
plt.title("Распределение веса новорожденных")
plt.xlabel("Вес (унции)")
plt.ylabel("Плотность")
plt.legend()
plt.show()
```

Расчет характеристик распределения, близкого к нормальному

```python
# 1. Математическое ожидание (среднее значение)
mean_value_1 = weights.mean()
print(f"Математическое ожидание: {mean_value_1:.3f} унций")

# 2. Дисперсия
variance_1 = weights.var()
print(f"Дисперсия: {variance_1:.3f} унций²")

# 3. Асимметрия
skewness_1 = weights.skew()
print(f"Асимметрия: {skewness_1:.3f}")

# 4. Эксцесс
kurtosis_1 = weights.kurtosis()
print(f"Эксцесс: {kurtosis_1:.3f}")

# 5. Квантили уровня 0,05 и 0,95
quantile_05_1 = weights.quantile(0.05)
quantile_95_1 = weights.quantile(0.95)
print(f"Квантиль 0,05: {quantile_05_1:.3f} унций")
print(f"Квантиль 0,95: {quantile_95_1:.3f} унций")

# 2,5%-ная точка
quantile_025_1 = weights.quantile(0.025)
print(f"2,5%-ная точка: {quantile_025_1:.3f} унций")
```
